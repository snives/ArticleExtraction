# Article Extraction using Neural Network
The purpose of this notebook is to investigate the efficacy of using a neural network to separating the article body of a web page from the boilerplate, advertising, and ancillary content.  

This is an important preprocessing step to increase the quality of subsequent text processing.  There are various methods of doing this with varying degrees of success.  The question we wish to answer is whether a neural network can perform this process with higher accuracy than current state of the art processes.

I first start with some feature engineering and creating a training dataset, with various columns which I hypothesize will have statistical power to separate text elements into article and non-article classes.  This was performed prior to this notebook and provided in `training.csv`.  I then inspect the dataset and decide to transform certain columns to make their distributions more normal using Box-Cox, followed by standardization or normalization, where applicable.

Categorical features such as `Tag` are one-hot encoded, and the `AncestorTag`, which is a Bitvector of the most informative 32 ancestor tags of the given tag, are simply encoded as 32 separate boolean features.

The dependent variable `Output` is biased, that is its classification is imbalanced and so I employ a Synthetic Minority Over-Sampling Technique to equalize the classes.

I use Keras with a Tensorflow backend to define a 3 layer neural network, using ReLu activation function and RMSProp optimizer, training for 50 epochs with batches of 100.  The neural network achieves a 98% accuracy at predicting the correct classification of each tag in a html document.

#Transform columns with non-normal distributions, to normal distributions using box-cox transform, visualized above.
data['TextLength'] = stats.boxcox(data['TextLength'].values, lmbda=text_maxlog)
data['Punctuation'] = stats.boxcox(data['Punctuation'].values + 1, lmbda=punc_maxlog)
data['NegativeKeywordCount'] = stats.boxcox(data['NegativeKeywordCount'].values + 1, lmbda=neg_maxlog)
data['PositiveClassCount'] = stats.boxcox(data['PositiveClassCount'].values + 1, lmbda=punc_maxlog)
data['TextToTagRatio'] = stats.boxcox(data['TextToTagRatio'].values, lmbda=ttr_maxlog)


```python
#Store the lambda values for these box-cox transforms
boxcox_transforms = {'TextLength': {'lambda':text_maxlog, 'offset': 0},
                     'Punctuation': {'lambda':punc_maxlog, 'offset': 1},
                     'NegativeKeywordRatio': {'lambda':neg_maxlog, 'offset': 1},
                     'PositiveClassCount' : {'lambda':punc_maxlog, 'offset': 1},
                     'TextToTagRatio' : {'lambda':ttr_maxlog, 'offset': 0},
                     'SimilarAncestorRatio' : {'lambda':sar_maxlog, 'offset': 1},
                    }
#Calculate lambdas
for (column, val) in boxcox_transforms.items():
    xt, lmbda = stats.boxcox(data[column].values + val['offset'])
    boxcox_transforms[column]['lambda'] = lmbda
```

```python
#Perform transforms
for (column, val) in boxcox_transforms.items():
        data[column] = stats.boxcox(data[column].values + val['offset'], lmbda=val['lambda'])

#Store these parameters for later
import joblib
joblib.dump(boxcox_transforms, 'lambdas.pkl')
```

```python
#Evaluate our classification distribution of values.  Is the data dominated by one classification?
plt.hist(Y_train)

#shows our data is predominantly false  852/1140 ~ 74% false, called "imbalanced".
#This means the NN could simply be biased negatively and get high accuracy without learning.

#The solution is to have a nearly unbiased output classification, e.g. 50% of each class.
#This can be done by underrepresenting the dominant classes, or overrepresenting the lesser classes.
#There is a python package that will do this called SMOTE (Synthetic Minority Over-Sampling Technique)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE('minority')

print(X_train.shape, Y_train.shape)
X_train, Y_train = smote.fit_sample(X_train, Y_train)
print(X_train.shape, Y_train.shape)
```

```python
#Verify the class distribution has been equalized
plt.hist(Y_train)
```

```python
#Now we can begin creating our model in keras
from keras import models
from keras import layers
from keras.callbacks import LearningRateScheduler

model = models.Sequential()
model.add(layers.Dense(80, activation='relu', use_bias=True, input_shape=(X_train.shape[1],)))
#model.add(layers.Dropout(0.05))
model.add(layers.Dense(32, activation='relu', use_bias=True, ))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, Y_train, epochs=50, batch_size=100, validation_data=(X_test, Y_test))

#For more optimizers see: https://towardsdatascience.com/learning-rate-schedules-and-adaptive-learning-rate-methods-for-deep-learning-2c8f433990d1
```

```python
#Lets evaluate the learning

history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

#history_dict
epochs = range(1, len(history_dict['loss']) + 1)


plt.plot(epochs, loss_values, 'bo', label='Training Loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#Display Accuracy
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']

plt.plot(epochs, acc_values, 'bo', label='Training Accuracy')
plt.plot(epochs, val_acc_values, 'b', label='Validation Accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
